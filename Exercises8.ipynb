#### **Exercise 1:**
Give students in the `example.csv` file a bonus of 5 for all the courses. Find a way to replace all the values that are now greater than 30 by 30. Save the new data to a file. Make a plot with histograms for every course in the example file.

```python
import pandas as pd
import matplotlib as plt
data = pd.read_csv('data/example.csv',index_col = 0)
#print(data)
#print(data.apply(lambda x:x+5))
data = data.apply(lambda x: x + 5)
#print(data)
data = data.clip(upper = 30)
#print(data)
data.to_csv('replaced_data.csv', index = False)
data.plot(kind = 'hist')
```
