#### **Exercise 1:**
Create a function returning the first N terms of the tailor series of the function without using the sympy series function.

```python
%pip install sympy   
```

```python
import sympy as sp

def taylor(f,x0,N):
    x = sp.symbols('x')
    temp = f(x)
    #temp = sp.Derivative(f(x),x, evaluate = True)
    fres = 0
    for i in range(N):
        fres += temp*(x-x0)**i/sp.factorial(i)
        temp = sp.Derivative(temp,x,evaluate = True)
        #print(temp)
    print(fres)
    return
def f(x):
    return sp.sin(x)
taylor(f,0,5)
```

```python
import time as t
import numpy as np

list1 = [[1,3,4],[5,6,7],[12,1,3]]
list2 = [[3,4,6],[1,3,6],[7,8,10]]
list3 = [[0,0,0],[0,0,0],[0,0,0]]
print(list1,list2)
start = t.process_time()
for i in range(len(list1)):
    for j in range(len(list2)):
        list3[i][j] = sum([x*y for x in list1[i] for y in list2[j]])
end = t.process_time()
print(f"Process time: {end - start} seconds")
l1 = np.array([1,3,4,5,6,7,10])
l2 = np.array([3,4,6,1,3,67,8])
start = t.process_time()
l3 = l1*l2
end = t.process_time()
print(f"Process time: {end - start} seconds")


```

#### **Exercise 4:**
Plot the recaman sequence in the following way:

Add a semi-circular arc from $a_{n}$ to $a_{n-1}$.

Alternate the arcs to be above and below the x-axis according to whether
    n is even or odd.

```python
import numpy as np
from matplotlib import pyplot as plt

def recaman(N):
    an = 0
    solution_list = [[0,0]]
    sol_l = [0]
    solution_set = set(sol_l)
    for n in range(1,N):
        an = sol_l[-1]
        condition = an - n
        if condition > 0 and condition not in solution_set:
            an = sol_l[-1] - n
        else:
            condition = sol_l[-1] + n
        solution_list.append([n,condition])
        sol_l.append(condition)
        solution_set.add(condition)
    print(solution_list)
    return sol_l
def graph(N):
    try:
        d = np.linspace(-1,1)
        y = np.array([0])
        x = np.array([0])
        l = recaman(N)
        for n in range(1,len(l)):
            xl = np.real(x[-1]) #Last point of x
            yl = np.real(y[-1]) #Last point of y
            r = (l[n]-l[n-1])/2
            yr = r*np.sqrt(1-d**2)
            x = np.append(x,xl + r*(d+1))
            y = np.append(y,yl + yr)
        plt.plot(x[1:],y[1:])
        plt.xlabel('x')
        plt.ylabel('y')
        plt.title(f'Up and down arcs for the first {N} Recaman numbers')
        plt.show()
    except Exception as e:
        print(f"Error:{e}")
    return
graph(75)
```

#### **Exercise 7:**

Repeat the curve fitting example (from SciPy lecture) with the following modifications:

 - choose some other function
 - first write the data to a file, then read it (with numpy, for example)
 - produce the picture that has axes, name and legend


```python
from scipy import optimize

np.random.seed(0)

# Let's generate fake data that we can "fit" then 
def f(x, a, phi):
    return np.sin(a * x + phi) + phi*x

#random makes it look like experimental data
x = np.linspace(0, 4, 50)
y = f(x, 1.3, 0.9) + .1*np.random.normal(size=50)
np.savetxt("data.txt",[x,y])

data = np.loadtxt("data.txt")
# Now fit the model: the parameters omega and phi can be found in the
# `params` vector
params, params_cov = optimize.curve_fit(f, data[0,:], data[1,:])

print(params)

# plot the data and the fitted curve
plt.plot(data[0,:],data[1,:], 'bx')
plt.plot(x,f(x,1.3,0.9),'g-')
plt.plot(x, f(x, *params), 'r-')
```

#### **Exercise 8:**

Integrate the function `x**2 + 10*np.sin(x)` on the interval `[-10,10]`.
Find a way to make a plot like in textbooks where you color the area under the curve that you integrate.
Now integrate the same function symbolically using `sympy`, then put borders into it and compare the results.

```python
import sympy as sp
def fint():
    x = np.linspace(-10,10,150)
    y = x**2 + 10*np.sin(x)
    tsum = 0
    for i in range(1,len(x)):
        tsum += y[i]*(x[i]-x[i-1])
    plt.plot(x,y,color = 'black')
    plt.fill_between(x,y,color = 'orange')
    plt.title('Integration')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    print(f"First integral result:{tsum}")
    return tsum
def fint2():
    try:
        x = sp.symbols('x')
        f = sp.simplify(x**2 + 10*sp.sin(x))
        print("Integral result:")
        return sp.integrate(f,(x,-10,10))
    except Exception as e:
        print(f"Error:{e}")
fint()
fint2()
```
